# Problem Statement
Write a program to implement Back-propagation algorithm using ANN.  Use suitbale datatset.

```python
import pandas as pd

import seaborn as sb
from matplotlib import pyplot as plot

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('./iris.csv')
df.head()
```

# Data Visualization

```python
plot.figure(1, (15, 7))
sb.lineplot(x='sepal.length', y='sepal.width', data=df, hue='variety')
```

```python
plot.figure(1, (15, 7))
sb.lineplot(x='petal.length', y='petal.width', data=df, hue='variety')
```

# Data Preprocessing

```python
x = df.drop('variety', axis=1)
y = df['variety']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
```

# Model Building

```python
mlpmodel = MLPClassifier(
    hidden_layer_sizes=(20, 10, 30),
    activation='relu', solver='adam'
)
```

```python
mlpmodel.fit(x_train, y_train)
```

```python
predictions = mlpmodel.predict(x_test)
```

# Metrics

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, predictions)
```

```python
plot.figure(1, (15, 7))
sb.heatmap(confusion_matrix(y_test, predictions), annot=True)
```

```python
print(classification_report(y_test, predictions))
```

```python
score = accuracy_score(y_test, predictions)
print("The accuracy is: ", score)
```
